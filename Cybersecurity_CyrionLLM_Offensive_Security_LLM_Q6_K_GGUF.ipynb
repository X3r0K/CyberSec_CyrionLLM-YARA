```python
predictions = []

# Limit to the first 5 predictions to avoid excessive runtime and potential KeyboardInterrupt
# llama-cpp-python should automatically use the GPU if available.
for i, item in enumerate(test_dataset[:5]):
    # Construct the full prompt for each test item
    full_prompt_test = [
        {"role": "system", "content": system_prompt}
    ]
    for example in few_shot_examples:
        full_prompt_test.extend(example["messages"])

    # Add the current test item's user message
    full_prompt_test.append(item["messages"][0]) # Assuming the first message is the user's instruction

    # Generate a response using the LLM
    response = llm.create_chat_completion(
        messages=full_prompt_test,
        max_tokens=500 # Ensure enough tokens for a complete response
    )

    # Extract and store the prediction
    predictions.append(response["choices"][0]["message"]["content"])

print(f"Generated {len(predictions)} predictions for the test dataset.")
```
